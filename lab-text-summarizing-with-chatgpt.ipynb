````python
from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv

# Read the local .env file if it exists.
# The .env file should contain this line:
# OPENAI_API_KEY=your_private_key_here
#
# IMPORTANT:
# Do not upload your .env file to GitHub.
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# I use a current, lightweight model for the lab.
# If the teacher asks for a different model, I only change this line.
MODEL_NAME = os.getenv("OPENAI_MODEL", "gpt-4o-mini")

if OPENAI_API_KEY is None:
    print("API key not found. Create a .env file with OPENAI_API_KEY=your_key_here before running API calls.")
else:
    print("API key loaded: True")
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product \ 
    review from an ecommerce site. 

    Summarize the review below, delimited by triple \
    backticks in at most 20 words. 

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````

````python
# VERSION 1: General short summaries
# This prompt asks for a normal summary of each review.
# It is useful when we only want the main idea of the review.

print("VERSION 1: GENERAL SHORT SUMMARIES\n")

for i, review in enumerate(reviews, start=1):
    prompt = f"""
    Your task is to summarize a product review from an ecommerce site.

    Summarize the review below in at most 25 words.
    Keep only the most important idea.

    Review: ```{review}```
    """

    response = get_completion(prompt)
    print(f"Review {i}:")
    print(response)
    print()
````

````python
# VERSION 2: Department-focused summaries
# This prompt is more specific.
# It creates summaries for three departments: Shipping, Pricing and Customer Support.
# This is useful because different teams care about different information.

print("VERSION 2: DEPARTMENT-FOCUSED SUMMARIES\n")

for i, review in enumerate(reviews, start=1):
    prompt = f"""
    You are analyzing ecommerce product reviews for internal company teams.

    Read the review below and write three short bullet points:
    - Shipping: mention only delivery, arrival time, transit damage or missing items.
    - Pricing: mention only price, value, discounts or expensive parts.
    - Customer Support: mention only support, warranty, replacement or service experience.

    If the review does not mention a topic, write "Not mentioned".

    Review: ```{review}```
    """

    response = get_completion(prompt)
    print(f"Review {i}:")
    print(response)
    print()
````
